## Warm-Up Exercise


Before you attempt the actual lab, here is a fully solved warmup exercise that will help you to learn how to access an API.


Using an API, let us find out who currently are on the International Space Station (ISS). The API at [http://api.open-notify.org/astros.json](http://api.open-notify.org/astros.json) gives us the information of astronauts currently on ISS in json format.


```python
import requests  # you need this module to make an API call
import pandas as pd
```

```python
api_url = "http://api.open-notify.org/astros.json"  # this url gives us the astronaut data
```

```python
response = requests.get(api_url)  # Call the API using the get method and store the
                                   # output of the API call in a variable called response.
```

```python
if response.ok:             # if all is well (no errors, no network timeouts)
    data = response.json()  # store the result in json format in a variable called data
                            # the variable data is of type dictionary.
```

```python
print(data)   # print the data just to check the output or for debugging
```

Print the number of astronauts currently on ISS.


```python
print(data.get('number'))
```

Print the names of the astronauts currently on ISS.


```python
astronauts = data.get('people')
print("There are {} astronauts on ISS".format(len(astronauts)))
print("And their names are :")
for astronaut in astronauts:
    print(astronaut.get('name'))
```

## Lab: Collect Jobs Data using Jobs API


### Objective: Determine the number of jobs currently open for various technologies and for various locations


Collect the number of job postings for the following locations using the API:

* Los Angeles
* New York
* San Francisco
* Washington DC
* Seattle
* Austin
* Detroit


```python
# Import required libraries
import requests
import pandas as pd
import json
```

#### Load the Jobs Dataset

> **Note:** The Jobs API runs as a local Flask server (`http://127.0.0.1:5000/data`). 
> In this notebook we load the same dataset directly from its public URL so everything works without running Flask.
> If you have the Flask server running locally, replace `load_jobs_data()` calls with `requests.get(api_url, params=...)` as shown in the comments.


```python
# ── Load dataset from the public URL (same data the Flask API serves) ──
JOBS_URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/module%201/Accessing%20Data%20Using%20APIs/jobs.json"

response = requests.get(JOBS_URL)
jobs_data = response.json()  # list of job dictionaries

print(f"Total job records loaded: {len(jobs_data)}")
print("Sample record keys:", list(jobs_data[0].keys()))
print("\nSample record:")
print(jobs_data[0])
```

#### Write a function to get the number of jobs for a given Technology

> **Keys in the JSON:** `Job Title`, `Job Experience Required`, `Key Skills`, `Role Category`, `Location`, `Functional Area`, `Industry`, `Role`


```python
api_url = "http://127.0.0.1:5000/data"  # local Flask API endpoint

def get_number_of_jobs_T(technology):
    """
    Returns the number of job postings whose 'Key Skills' field
    contains the given technology keyword (case-insensitive).
    
    When the Flask server is running locally you can replace the body with:
        payload = {"Key Skills": technology}
        response = requests.get(api_url, params=payload)
        data = response.json()
        number_of_jobs = len(data)
    """
    number_of_jobs = sum(
        1 for job in jobs_data
        if technology.lower() in job.get("Key Skills", "").lower()
    )
    return technology, number_of_jobs
```

Calling the function for Python and checking if it works.


```python
get_number_of_jobs_T("Python")
```

#### Write a function to find number of jobs in US for a given Location


```python
def get_number_of_jobs_L(location):
    """
    Returns the number of job postings whose 'Location' field
    contains the given location keyword (case-insensitive).
    
    When the Flask server is running locally you can replace the body with:
        payload = {"Location": location}
        response = requests.get(api_url, params=payload)
        data = response.json()
        number_of_jobs = len(data)
    """
    number_of_jobs = sum(
        1 for job in jobs_data
        if location.lower() in job.get("Location", "").lower()
    )
    return location, number_of_jobs
```

Call the function for Los Angeles and check if it is working.


```python
get_number_of_jobs_L("Los Angeles")
```

### Store the results in an Excel file


#### Create a python list of all locations for which you need to find the number of job postings.


```python
locations = [
    "Los Angeles",
    "New York",
    "San Francisco",
    "Washington DC",
    "Seattle",
    "Austin",
    "Detroit"
]
```

#### Import libraries required to create an Excel spreadsheet


```python
from openpyxl import Workbook
```

#### Create a workbook and select the active worksheet


```python
wb = Workbook()          # create a new workbook
ws = wb.active           # use the active (default) worksheet
ws.title = "Jobs by Location"

# Add header row
ws.append(["Location", "Number of Jobs"])
```

#### Find the number of job postings for each location and write to the spreadsheet


```python
for location in locations:
    loc, count = get_number_of_jobs_L(location)
    ws.append([loc, count])
    print(f"{loc}: {count} jobs")
```

#### Save into an Excel spreadsheet named `job-postings.xlsx`


```python
wb.save("job-postings.xlsx")
print("Saved: job-postings.xlsx")
```

#### Collect the number of job postings for the following technologies and store in the same Excel file

Technologies to check:
* C
* C#
* C++
* Java
* JavaScript
* Python
* Scala
* Oracle
* SQL Server
* MySQL Server
* PostgreSQL
* MongoDB


```python
technologies = [
    "C", "C#", "C++", "Java", "JavaScript", "Python",
    "Scala", "Oracle", "SQL Server", "MySQL Server",
    "PostgreSQL", "MongoDB"
]
```

```python
# Load the workbook we saved earlier and add a new sheet for technologies
from openpyxl import load_workbook

wb = load_workbook("job-postings.xlsx")

# Create a new sheet for technology data
ws_tech = wb.create_sheet(title="Jobs by Technology")

# Add header row
ws_tech.append(["Technology", "Number of Jobs"])

# Write technology job counts
for tech in technologies:
    technology, count = get_number_of_jobs_T(tech)
    ws_tech.append([technology, count])
    print(f"{technology}: {count} jobs")

# Save the updated workbook
wb.save("job-postings.xlsx")
print("\nUpdated workbook saved: job-postings.xlsx")
```

#### Preview the results as DataFrames


```python
# Preview Location results
df_locations = pd.read_excel("job-postings.xlsx", sheet_name="Jobs by Location")
print("=== Jobs by Location ===")
print(df_locations.to_string(index=False))
```

```python
# Preview Technology results
df_technologies = pd.read_excel("job-postings.xlsx", sheet_name="Jobs by Technology")
print("=== Jobs by Technology ===")
print(df_technologies.to_string(index=False))
```
